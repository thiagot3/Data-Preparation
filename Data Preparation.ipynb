```python
# \d RegEx para identificar nÃºmeros de 0 a 9
df.launched = df.launched.str.replace(' \d\d:\d\d', '', regex=True)   
# hora formatada, porem continua como objeto
print(df.dtypes)
df.head()
```

```python
df.head()
```

```python
# mudando tipo das colunas para datetime
df['deadline'] = pd.to_datetime(df.deadline, format = '%d/%m/%y')
df['launched'] = pd.to_datetime(df.launched, format = '%d/%m/%y')
print(df.dtypes)
df.head()
```

```python
df["time range in days"] = (df.deadline - df.launched).dt.days
df
```

```python
invested_df = pd.read_csv(r'Arquivos da AplicaÃ§Ã£o/invested.csv')
invested_df.head()
```

```python
invested_df = pd.get_dummies(invested_df, columns = ['backedLocation'])
invested_df
```

```python
invested_df = invested_df.groupby('ID').agg({'age': 'mean', 'backedLocation_BR' : 'sum', 'backedLocation_GBK': 'sum', 'backedLocation_US' : 'sum'})
invested_df.head()
```

```python
df = df.merge(invested_df, how = 'right', on = 'ID')
df.head()
```
